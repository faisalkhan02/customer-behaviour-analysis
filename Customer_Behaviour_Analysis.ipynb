## 🧹 Step 4 — Data Cleaning

### 4.1 Fix Category Mismatch

The `Category` column has incorrect category assignments for some items.

```python
# Check all unique values in 'Category' column
# Some items may be incorrectly categorised in the original dataset
df['Category'].unique()
```

### 4.2 Handle Null Values in `Size` Column

The `Size` column has nulls introduced due to category remapping.  
Different categories need different handling strategies:
- **Electronics** → Size is not applicable (no clothing size needed)
- **Accessories** → One-size-fits-all (Free Size)
- **Footwear** → Size info not available in this dataset
- **Clothing** → Fill with the most common size (mode)

```python
# Check all unique size values currently in the dataset
df['Size'].unique()
```

```python
# Inspect which sizes exist per category
# This confirms why we need a category-specific strategy
df.groupby('Category')['Size'].unique()
```

```python
# See which sizes are used per item — further granular view
df.groupby('Item Purchased')['Size'].unique()
```

```python
# Electronics don't require clothing sizes
# Replacing all Size values for Electronics with 'Not Applicable'
df.loc[df['Category'] == 'Electronics', 'Size'] = "Not Applicable"
```

```python
# Verify Electronics category now has 'Not Applicable' in Size
df.groupby('Category')['Size'].unique()
```

```python
# Accessories are one-size-fits-all items (bags, belts, scarves, etc.)
# Replacing all Size values for Accessories with 'Free Size'
df.loc[df['Category'] == 'Accessories', 'Size'] = "Free Size"
```

```python
# Verify Accessories category now has 'Free Size'
df.groupby('Category')['Size'].unique()
```

```python
# Footwear size data is not available in this dataset
# Replacing all Size values for Footwear with 'Not Available'
df.loc[df['Category'] == 'Footwear', 'Size'] = "Not Available"
```

```python
# Verify Footwear category now shows 'Not Available'
df.groupby('Category')['Size'].unique()
```

```python
# For Clothing, fill only the null sizes (not all values)
# Strategy: use the most frequent (mode) size within Clothing category
# mode()[0] gives the single most common size value
df.loc[(df['Category'] == 'Clothing') & (df['Size'].isnull()), 'Size'] = \
    df[df['Category'] == 'Clothing']['Size'].mode()[0]
```

```python
# Verify no nulls remain in Size for Clothing
df.groupby('Category')['Size'].unique()
```

```python
# Final check — confirm all Size nulls are resolved
df.isnull().sum()
```

---
### 4.3 Handle Null Values in `Review Rating`

**Strategy:** Fill missing ratings with the **mean rating of that specific item**.  
This is better than a global mean because different items naturally have different average ratings.

```python
# Preview current state of the data
df.head()
```

```python
# Calculate overall mean review rating across all items
df["Review Rating"].mean()
```

```python
# Calculate overall median review rating
df["Review Rating"].median()
```

```python
# Find the most frequently occurring review rating (mode)
df["Review Rating"].mode()
```

```python
# Calculate mean review rating grouped by each item
# This is the value we will use to fill nulls (item-specific mean)
df.groupby("Item Purchased")["Review Rating"].mean()
```

```python
# Fill null Review Rating values with the mean rating of the same item
# transform('mean') aligns the group mean back to each row's index
df["Review Rating"] = df["Review Rating"].fillna(
    df.groupby("Item Purchased")["Review Rating"].transform('mean')
)
```

```python
# Verify Review Rating nulls are resolved
df.isnull().sum()
```

```python
# Preview the cleaned data
df.head()
```

---
### 4.4 Handle Null Values in `Previous Purchases`

**Strategy:** Fill missing values with **0**.  
Rationale: If a customer has no previous purchase history, it's safe to assume 0 prior purchases.

```python
# Fill null 'Previous Purchases' values with 0
# Assumption: null means no recorded previous purchases
df['Previous Purchases'] = df['Previous Purchases'].fillna(0)
```

```python
# Confirm no nulls remain in Previous Purchases
df.isnull().sum()
```

---
### 4.5 Handle Null Values in `Purchase Amount (USD)`

**Strategy:** Fill missing amounts with the **mean purchase amount of the same item**.  
This preserves realistic price ranges per product type.

```python
# View mean purchase amount for each item — used for filling nulls
df.groupby("Item Purchased")["Purchase Amount (USD)"].mean()
```

```python
# Fill null Purchase Amount with item-specific mean price
# This ensures price imputation is realistic and item-appropriate
df["Purchase Amount (USD)"] = df["Purchase Amount (USD)"].fillna(
    df.groupby("Item Purchased")["Purchase Amount (USD)"].transform('mean')
)
```

```python
# Final null check — all columns should now show 0 nulls
df.isnull().sum()
```

## 🔁 Step 5 — Handle Duplicate Records

Check for duplicate rows. Since each row represents a unique customer transaction,

## ✏️ Step 6 — Standardise Column Names

Rename columns to **snake_case** (lowercase with underscores).  
This ensures compatibility with SQL databases and Power BI without quoting or escaping column names.

```python
# Step 1: Replace spaces with underscores in all column names
# Step 2: Convert all column names to lowercase
# Result: 'Item Purchased' → 'item_purchased', 'Review Rating' → 'review_rating'
df.columns = df.columns.str.replace(" ", "_")
df.columns = df.columns.str.lower()
```

```python
# Verify the renamed columns
df.columns
```

```python
# 'purchase_amount_(usd)' has brackets which are problematic in SQL
# Rename it to a clean 'purchase_amount' column name
df = df.rename(columns={"purchase_amount_(usd)": "purchase_amount"})
```

```python
# Final column list — all clean, lowercase, snake_case names
df.columns
```
